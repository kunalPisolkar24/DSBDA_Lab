```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv")
df
```

```python
df.head()
```

```python
x = df.drop("medv", axis = 1)

#output data
y = df["medv"]
```

```python
print("Shape of medv: ", x.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 0, test_size = 0.25)
x_train
```

```python
x_train.head()
```

```python
print("shape of train data: ", x_train.shape)
print("Shape of the test data: ", x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train, y_train)
print("Regression coeff: \n", reg.coef_)
print("\n Regression intercept: \n", reg.intercept_)
```

```python
# predictions
y_pred = reg.predict(x_test)
print("Shape of y_pred: ", y_pred.shape)
```

```python
result = pd.DataFrame({"Actual": y_test, "Produced": y_pred})
print("Comparing Results: \n\n", result)
```

```python
residual_err = abs(y_test - y_pred)
print("Residual Error \n\n", residual_err)
print("\n \n Mean Absolute Error: ", sum(residual_err) / len(residual_err))

from sklearn.metrics import mean_absolute_error
meanAbs = mean_absolute_error(y_pred, y_test)
print("\n\nMean Absolute Error from SkLearn: ", meanAbs)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
print("Mean absolute percentage error: ", mean_absolute_percentage_error(y_test, y_pred))
```

```python
from sklearn.metrics import r2_score
print("R2 Score from SK_learn: ", r2_score(y_test, y_pred))
```
