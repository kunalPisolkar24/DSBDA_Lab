```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns. load_dataset('titanic')
df= df[['sex','age','survived']]
print(df.head())
print("\n\n", df.dtypes)
```

```python
# Create a new figure and a set of subplots
plt.figure(figsize=(10, 6))

# Create a box plot for 'sex' and 'age' with a custom color palette
sns.boxplot(x='sex', y='age', data=df, palette='coolwarm')

# Set the title of the box plot
plt.title('Box plot of Age by Sex')

# Set the x and y axis labels
plt.xlabel('Sex')
plt.ylabel('Age')

# Display the box plot
plt.show()
```

```python
# Create a new figure and a set of subplots
plt.figure(figsize=(10, 6))

# Create a box plot for 'sex' and 'age' with 'survived' as hue and a custom color palette
sns.boxplot(x='sex', y='age', hue='survived', data=df, palette='coolwarm')

# Set the title of the box plot
plt.title('Box plot of Age by Sex with Survived as hue')

# Set the x and y axis labels
plt.xlabel('Sex')
plt.ylabel('Age')

# Display the box plot
plt.show()
```
