```python
import numpy as np
print("Standard Deviation of marks: ", np.std(df['marks']))
```

```python
# Group by class and TE
gr1 = df.groupby('class')
te = gr1.get_group('TE')
```

```python
print("Min of TE: \n", te.min())
```

```python
print("Max of TE: \n", te.max())
```

```python
# Grouping by age
gr2 = df.groupby('age')
print("Grouping by age: ", gr2.groups)
```

```python
import seaborn as sns
# we can load the iris example dataset from seaborn
df1 = sns.load_dataset('iris')
df1
```
